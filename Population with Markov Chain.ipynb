```python
import pandas as pd

pd.set_option('precision',5)

trans_2013 = pd.read_excel('in-out.xlsx', '2013', index_col='in/out', na_values=['NA'])
trans_2012 = pd.read_excel('in-out.xlsx', '2012', index_col='in/out', na_values=['NA'])
trans_2011 = pd.read_excel('in-out.xlsx', '2011', index_col='in/out', na_values=['NA'])

trans_2013.iloc[:, :] = trans_2013.iloc[:, :].astype(float)/trans_2013.iloc[:, :].sum()
trans_2012.iloc[:, :] = trans_2012.iloc[:, :].astype(float)/trans_2012.iloc[:, :].sum()
trans_2011.iloc[:, :] = trans_2011.iloc[:, :].astype(float)/trans_2011.iloc[:, :].sum()

# print trans_2013.iloc[1, :].sum()
# print trans_2013.iloc[2, :].sum()
# print trans_2013.iloc[3, :].sum()

```
